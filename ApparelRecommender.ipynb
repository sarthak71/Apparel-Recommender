```
from surprise.model_selection import train_test_split
from surprise import Dataset
from surprise import Reader

reader = Reader(rating_scale=(new_df['rating'].min(), new_df['rating'].max()))
data = Dataset.load_from_df(new_df[['visitorid', 'itemid', 'rating']],reader=reader)

trainset, testset = train_test_split(data, random_state=1, test_size=0.2)
```

```
from surprise import KNNWithMeans
from surprise import SVD
from surprise import accuracy
```

```
#train the model
#try with other hyperparams or k

#default k
algo1 = KNNWithMeans(sim_options={'name': 'pearson_baseline', 'user_based': True})
algo1.fit(trainset)
#custom k
algo2 = KNNWithMeans(k=101, sim_options={'name': 'pearson_baseline', 'user_based': True})
algo2.fit(trainset)
#svd algorithm
algo3 = SVD()
algo3.fit(trainset)
```

```
#Collaborative filtering model
# run the trained model against the testset
test_pred = algo1.test(testset)
test_pred2 = algo2.test(testset)
test_pred3 = algo3.test(testset)
```

```
test_pred[0]
```

```
# get RMSE
print("User-based Model : Test Set")
accuracy.rmse(test_pred, verbose=True)
accuracy.rmse(test_pred2, verbose=True)
accuracy.rmse(test_pred3, verbose=True)
```

```
#Lets get top 5 recommendations

from collections import defaultdict

def get_top_n(predictions, n=5):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]
    return top_n
```

```
top_n = get_top_n(test_pred3, n=5)
recommendations_df = pd.DataFrame(columns=['userid', 'items'])
for uid, user_ratings in top_n.items():
    recommendations_df = recommendations_df.append({'userid':uid, 'items':[iid for (iid, _) in user_ratings]},ignore_index=True)
    #print('Recommendations for user', uid, ' are items', [iid for (iid, _) in user_ratings])
```
